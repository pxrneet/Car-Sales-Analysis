```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mariasaif/car-sales")

print("Path to dataset files:", path)
```

```python
print(os.listdir(path))
```

```python
df =pd.read_csv(path+"/Car_sales.csv")
```

```python
for i in df.columns:
  print(i,":",df[i].unique())
```

```python
plt.figure()
sns.barplot(x='Manufacturer', y='Sales_in_thousands', data=df)
plt.xticks(rotation=90)
plt.title("Sales by Manufacturer")
plt.show()
-+
```

```python
plt.figure()
sns.scatterplot(x='Horsepower', y='Price_in_thousands', data=df)
plt.title("Price vs Horsepower")
plt.show()
```

```python
plt.figure()
sns.histplot(df['Price_in_thousands'], bins=20)
plt.title("Price Distribution")
plt.show()


```

```python
plt.figure()
sns.boxplot(x='Vehicle_type', y='Fuel_efficiency', data=df)
plt.title("Fuel Efficiency by Vehicle Type")
plt.show()
```

```python
plt.figure()

corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure()
sns.lineplot(x='Model', y='Sales_in_thousands', data=df)
plt.xticks(rotation=90)
plt.title("Sales Trend by Model")
plt.show()
```

```python
plt.figure()
sns.countplot(x='Manufacturer', data=df)
plt.xticks(rotation=90)
plt.title("Number of Models by Manufacturer")
plt.show()
```
